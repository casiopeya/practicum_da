<font size="4"><b>О проекте:</b></font>

TED (от англ. technology, education, design — «технологии, образование, дизайн») — некоммерческий фонд, который проводит популярные конференции. На них выступают специалисты из разных областей и читают лекции на актуальные социальные, культурные и научные темы.

В этом проекте будет исследована история TED-конференций с помощью дашбодов в Tableau.

***Ссылка на построенные дашборды: https://public.tableau.com/views/ConferencesTEDv2/TED?:language=en-US&publish=yes&:display_count=n&:origin=viz_share_link***

<font size="4"><b>Описание данных:</b></font>

**Три датасета:** 

* **[tableau_project_data_1.csv](https://code.s3.yandex.net/data-analyst/Tableu/tableau_project_data_1.csv),**
* **[tableau_project_data_2.csv](https://code.s3.yandex.net/data-analyst/Tableu/tableau_project_data_2.csv),**
* **[tableau_project_data_3.csv](https://code.s3.yandex.net/data-analyst/Tableu/tableau_project_data_3.csv)**

**хранят данные о выступлениях и имеют одинаковую структуру:**


- `talk_id` — идентификатор выступления;
- `url` — ссылка на запись выступления;
- `title` — название выступления;
- `description` — краткое описание;
- `film_date` — дата записи выступления;
- `duration` — длительность в секундах;
- `views` — количество просмотров;
- `main_tag` — основная категория, к которой относится выступление;
- `speaker_id` — уникальный идентификатор автора выступления;
- `laughter_count` — количество раз, когда аудитория смеялась в ходе выступления;
- `applause_count` — количество раз, когда аудитория аплодировала в ходе выступления;
- `language` — язык, на котором велось выступление;
- `event_id` — уникальный идентификатор конференции.

**Датасет [tableau_project_event_dict.csv](https://code.s3.yandex.net/data-analyst/Tableu/tableau_project_event_dict.csv) - справочник конференций:**

- `conf_id` — уникальный идентификатор конференции;
- `event` — название конференции;
- `country` — страна проведения конференции.

**Датасет [tableau_project_speakers_dict.csv](https://code.s3.yandex.net/data-analyst/Tableu/tableau_project_speakers_dict.csv) - справочник авторов выступления:**

- `author_id` — уникальный идентификатор автора выступления;
- `speaker_name` — имя автора;
- `speaker_occupation` — профессиональная область автора;
- `speaker_description` — описание профессиональной деятельности автора.

## Дашборд "История выступлений"

***На дашбодре "История выступлений" представлена визуализация с распределением:***
- выступлений по странам (доля выступлений), 
- по годам (абсолютное количество выступлений),
- наиболее популярные тематики по годам (доля выступлений).

***По наблюдаемым данным можно отметить следующее:***

1. За всё время наблюдений больше всего выступлений было проведено в *США, Канаде и Великобритании.*
2. Самыми популярными категориями выступлений являются: *Наука (science), Технологии (technology), Культура (culture).*
3. Популярность категорий Наука и Культура спадает со временем, их опережают категории: Социум, Технологии и Глобальные проблемы. Последние 3 темы являются более приближенными к реальности, вероятно, имеют более обширную целевую аудиторию, они напрямую связаны с общим развитием человечества и последствиями этого. Кроме того, категории Наука и Культура более общны и изначально содержат в себе многие другие категории, поэтому, возможно, изменения со временм связаны не с падением популярности, а с дроблением начальных Науки и Культуры на более узкие тематики.

## Дашборд "Тематики выступлений"

**На дашборде "Тематики выступлений" представлена визуализация:**

- распределения выступлений по категориям,
- наиболее популярные тематики по странам,
- реакции аудитории в зависимости от тематики,
- распределение количества просмотров выступлений в зависимости от длительности выступлений.

***По наблюдаемым данным можно отметить следующее:***

1. Наиболее популярны категории Наука (science), Технологии (technology), Культура (culture).
2. Различия в популярности категорий между странами практически отсутствуют. Для всех стран самой популярной тематикой является Наука, далее - Технологии и Культура. Различия между странами появляются для в целом менее популярных категорий, тем не менее различия являются слабыми.
3. Чаще всего аудитория аплодирует на выступлениях, связанных с категориями Развлечения, Социальные изменения, Образования.
4. Чаще всего аудитория смеется на выступлениях, относящихся к категориям: Образование, Бизнес, Дизайн и Развлечения, реже всего - Социальные изменения и Глобальные проблемы. Самое "весёлое" выступление - *A Saudi, an Indian and Iranian walk into a Qatar bar ...*.
5. Линейной зависимости между длительностью выступления и количеством просмотров нет, однако в целом выступления длительностью более, чем 20 мин не обладают высоким количеством просмотров. Самым популярным (боолее 72 млн просмотров) оказалось выступление *Do schools kill creativity?*, самым длинным (около 62 мин) - *3 secrets to Netflix's success*.

## Дашборд "Авторы выступлений"

**На дашборде "Тематики выступлений" представлена визуализация:**

- с наиболее популярными областями деятельностями авторов,
- распределения количества выступлений, приходящегося на одного автора,
- списка авторов, сортированного по числу их выступлений,
- списка выступлений, сортированного по числу просмотров.

***По наблюдаемым данным можно отметить следующее:***

1. Большинство авторов являются Писателями, Художниками, Журналистами, Авторами и Дизайнерами.
2. Один автор обычно выступает один раз, однако есть и исключения. Самым частым спикером TED является Ганс Рослинг (Hans Rosling): он выступал на платформе 10 раз.
3. Среди дизайнеров чаще всего выступал Том Вуйец (Tom Wujec) - 4 раза. Самым популярным его выступлением оказалось *Build a tower, build a team* - более 7 млн просмотров.

## Дашборд "Конференции по проектам"

**На дашборде "Тематики выступлений" представлена визуализация:**

- количества выступлений по отдельным проектам TED,
- динамика распределения выступлений по проектам по годам,
- количество просмотров выступлений в зависимости от проектов,
- количество проведенных мероприятий в рамках проекта.

***По наблюдаемым данным можно отметить следующее:***

1. Больше всего мероприятий было организовано в рамках проекта TEDx.
2. Больше всего выступлений проведено в основном проекте TED, вторым по количеству выступлений является проект TEDx. Следующие по популярности - TED woman и TED Global.
3. В среднем, больше всего просмотров у выступлений проекта TED Talks.
